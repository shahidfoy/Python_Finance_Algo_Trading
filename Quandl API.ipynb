## Commodities

```python
df_gold = quandl.get('LBMA/GOLD', start_date=start, end_date=end)
```

```python
df_gold
```

```python
graph = go.Figure()
```

```python
graph.add_trace(go.Scatter(x=df_gold.index, y=df_gold['USD (AM)'], name='USD'))
graph.add_trace(go.Scatter(x=df_gold.index, y=df_gold['GBP (AM)'], name='GBP'))
graph.update_xaxes(
    rangeslider_visible=True, title='Slider above'
)
graph.show()
```

## Real Estates data

```python
# New York real estate data
midtown = quandl.get('ZILLOW/N466_MRPST', start_date=start, end_date=end)
harlem = quandl.get('ZILLOW/N27_MRPST', start_date=start, end_date=end)
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=midtown.index, y=midtown['Value'], name='NY Midtown'))
graph.add_trace(go.Scatter(x=harlem.index, y=harlem['Value'], name='NY Harlem'))
graph.update_xaxes(
    rangeslider_visible=True,title='New York Real Estate'
)
graph.show()
```
